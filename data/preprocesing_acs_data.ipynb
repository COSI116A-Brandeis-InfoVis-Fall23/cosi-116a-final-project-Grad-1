```python
import pandas as pd
import glob
import os

# Path to the folder containing the CSV files
folder_path = 'replace_it_with_your_folder_path'

# Use glob to find all CSV files in the folder
csv_files = glob.glob(folder_path + '*.csv')

# Sort the list of csv files based on the year extracted from the filename
csv_files_sorted = sorted(csv_files, key=lambda x: int(os.path.basename(x).split('.')[0]))

# Initialize an empty list to store the data
data_list = []

# Process each file in sorted order
for file in csv_files_sorted:
    # Extract the year from the filename
    year = os.path.basename(file).split('.')[0]

    # Read the CSV file
    df = pd.read_csv(file)

    # Add a new 'Year' column with empty strings
    df['Year'] = ""

    # Find rows with 'Total:' in the first column (assuming the first column contains the 'Total:' entries)
    total_rows = df[df.iloc[:, 0] == 'Total:'].index

    # Set the 'Year' for these rows to the filename (year)
    df.loc[total_rows, 'Year'] = year

    # Append to the list
    data_list.append(df)

# Concatenate all dataframes
all_data = pd.concat(data_list)

# Drop a specific column if needed
all_data = all_data.drop('United States!!Margin of Error', axis=1)


# Save the combined data to a new CSV file
all_data.to_csv("ACS_GROSS_RENT_AS_A_PERCENTAGE_OF_HOUSEHOLD_INCOME_11-22.csv", index=False)

all_data
```
